```python
# 算出订单时间距今多久（这里采用减去订单的最大时间），并将时间类型数据转化为数字，方便后续计算
rfm['R']=-(rfm.dt-rfm.dt.max())/np.timedelta64(1,'D') 
# 重命名
rfm.rename(columns={'products':'F','amount':'M'},inplace=True)
rfm.head()
```

```python
rfm[['R','F','M']].apply(lambda x: x-x.mean())  # 目的是将用户划分为低于平均维度与高于平均维度
```

```python
def rfm_func(x):
    level = x.apply(lambda x:'1' if x>=0 else '0')
    label = level.R+level.F+level.M
    d = {
        '111':'重要价值客户', # 'R'>0表示距今时间是较久远的（有一段时间没有消费了，可以发展），'F'>0表示产品消费数量较好，‘M’>0表示消费金额是较可观的
        '011':'重要保持客户', # 'R'<0表示消费时间挺近的，
        '101':'重要挽留客户',
        '001':'重要发展客户',
        '110':'一般价值客户',
        '010':'一般保持客户',
        '100':'一般挽留客户',
        '000':'一般发展客户'
    }
    result=d[label]
    return result
rfm['label']=rfm[['R','F','M']].apply(lambda x: x-x.mean()).apply(rfm_func,axis=1)
```

```python
从RFM分层可知，大部分用户为重要保持客户，但是这是由于极值的影响（从下面散点图也可以看到），所以RFM的划分标准应该以业务为准
1. 尽量用小部分的用户覆盖大部分的额度
2. 不要为了数据好看划分等级
3. 从下面散点图也可以看到，从F维度上可以看到极值比较夸张，整个散点图被F维度大于200的值拉伸开来了。
   极值会极大地影响均值，会把平均值给拉高或低，所以这种情况下，“x-x.mean()”会造成数据的偏离，不能真实反映数据的特征，
   #可以人工将mean()改为中位数或人工定义几个维度，比较灵活。或人工将极值过渡掉
```

### 4.2 新、老、活跃、回流、流失

- 新用户为第一次消费
- 活跃用户为在某个时间窗口内有过消费
- 不活跃/流失用户为在时间窗口内没有消费过的老客
- 回流用户为在上一个窗口中没有消费，而在当前时间窗口内有过消费

```python
# 查看用户在每个月的消费次数
pivoted_counts = df.pivot_table(
    index='id',
    columns='month',
    values='dt',
    aggfunc='count').fillna(0)
pivoted_counts.head()
```

```python
# 判断用户是否进行过消费
df_purchase=pivoted_counts.applymap(lambda x : 1 if x>0 else 0)
df_purchase.tail()
```

```python
df.loc[df.id==23566]
```

```python
# 上面是对所有数据进行了透视，从实际的业务场景来看，有些用户在1、2月份没有进行注册，3月份才是第一次消费，
# 如用户23566, 在原始数据中，对该用户的记录是从3月份开始的，没有1、2月份的记录，但数据透视会将1、2月份的补充上，填充为0，
# 所以要进行进一步操作，将第一次消费的时间作为生命周期的起点
```

```python
# unreg,表示未注册用户(registration)
# new, 表示新注册用户（第一次消费）
# active, 表示活跃用户，本月进行了消费，上个月也进行了消费
# unactive, 表示不活跃用户，本月没有消费，但之前消费过
# return, 表示回归用户，本月消费了，但上个月没有消费
def active_status(data):
    status = []
    # 行数据
    for i in range(18):
        # 若本月没有消费
        if data[i]==0:
            # 若之前有记录
            if len(status)>0:
                if status[i-1]=='unreg':
                    status.append('unreg')
                else:
                    status.append('unactive')
            # 若之前没有记录
            else:
                status.append('unreg')
        
        # 若本月消费了
        else:
            # 若之前没有记录
            if len(status)==0:
                status.append('new')
            # 若之前有记录
            else:
                if status[i-1]=='unactive':
                    status.append('return')
                elif status[i-1]=='unreg':
                    status.append('new')
                else:
                    status.append('active')
                    
    return pd.Series(status,index = df_purchase.columns)
```

若本月没有消费
- 若之前是未注册，则仍为未注册
- 若之前有消费，则为流失/不活跃用户
- （若没有之前）其他情况，为未注册

若本月有消费
- 若是第一次消费，则为新用户
- 如果之前有过消费，但上个月为不活跃，则为回流
- 如果上个有为未注册，则为新用户
- 除此外，为活跃

```python
purchase_stats = df_purchase.apply(active_status,axis=1)
purchase_stats.head()
```

```python
purchase_stats.tail()
```

```python
purchase_stats_ct = purchase_stats.replace('unreg',np.NaN).apply(lambda x:pd.value_counts(x))
purchase_stats_ct
# 得出每个月不同活跃程度用户的计数
```

```python
purchase_stats_ct.fillna(0).T.head()
```

```python
purchase_stats_ct.fillna(0).T.plot.area(figsize = (12,6))
```

```python
purchase_stats_ct.fillna(0).T.apply(lambda x: x/x.sum(),axis=1)
```

```python
#回流和活跃用户占比
return_rata = purchase_stats_ct.fillna(0).T.apply(lambda x: x/x.sum(),axis=1)
plt.figure(figsize = (12,4))
plt.subplot(121)
return_rata['return'].plot()
plt.subplot(122)
return_rata.active.plot()
```

## 4.复购率和回购率分析
- 复购率   
   自然月内、购买多次的用户占比
- 回购率   
   曾经购买过的用户在某一时期内的再次购买的占比
   

```python
pivoted_counts.head() # 用户在每个月内的消费次数
```

```python
purchase_r = pivoted_counts.applymap(lambda x: 1 if x>1 else np.NaN if x==0 else 0) # 匿名函数的嵌套写法
# 逻辑是，如果x>1则填写为1，剩下的情况中如果x=1则填写为0，如果为0则填写为空
purchase_r.head()
```

```python
# 复购率
(purchase_r.sum()/purchase_r.count()).plot(figsize=(10,4))
# purchase_r.count() 计算非空值的数量
```

复购率稳定在20%左右，前三个月因为有大量新用户涌入，而这批用户只购买了一次，所以导致复购率降低

```python
# 回购率，当月消费过次月仍在消费的人数/当月消费的总人数
df_purchase.head() # 1表示这个月消费过，0表示没有
```

```python
def purchase_back(data):
    status=[]
    for i in range(17):
        if data[i]==1:
            if data[i+1]==1:
                status.append(1)
            if data[i+1]==0:
                status.append(0)
        else:
            status.append(np.NaN)
    status.append(np.NaN)# 到数据最后一个月时，因为没有下一个月的数据，所以将其填为空值
    return pd.Series(status,index = df_purchase.columns)
```

```python
purchase_b = df_purchase.apply(purchase_back,axis=1)
purchase_b.head(5)
```

```python
# 回购率 
(purchase_b.sum()/purchase_b.count()).plot(figsize=(10,4))
```
